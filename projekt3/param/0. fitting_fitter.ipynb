```python
data = pd.read_csv(".\\data\\ropa.csv", index_col='Date', header=0)
data['returns'] = data.interpolate(type='linear').pct_change()
data.index = pd.to_datetime(data.index)
data
```

```python
fig = plt.figure(figsize = (10, 6))
plt.plot(data['returns'])
```

```python
sns.displot(data['returns'], kind="hist", kde=True)
```

```python
ceny = list(data['returns'])
ceny= ceny[1:]
```

```python
from fitter import Fitter
f = Fitter(ceny,timeout=360)
f.fit()
# may take some time since by default, all distributions are tried
# but you call manually provide a smaller set of distributions
```

```python
f.summary()
```

```python
a = f.summary(Nbest=np.inf, plot=False)
```

```python
len(f.distributions)
```

```python
distributions = a.index
SSE = list(a['sumsquare_error'])
AIC = list(a['aic'])
BIC = list(a['bic'])
```

```python
results_df = pd.DataFrame(np.transpose([distributions, SSE, AIC, BIC]))
results_df.columns = ['distributions', 'SSE', 'AIC', 'BIC']
results_df = results_df[:100]
```

```python
fitted_params = [f.fitted_param[i] for i in results_df['distributions']]
results_df['params'] = fitted_params
results_df.head(10)
```

```python
results_df.to_csv('fitted_distributions_best.csv')
```
